```python
import pandas as pd
import numpy as np
```

```python
fireforest=pd.read_csv("forestfires (2).csv")
```

```python
fireforest
```

```python
seed=7
```

```python
np.random.seed(seed)
```

```python
fireforest.info()
```

```python
from sklearn import preprocessing
```

```python
le=preprocessing.LabelEncoder()
```

```python
fireforest["day"]=le.fit_transform(fireforest["day"])
```

```python
fireforest["month"]=le.fit_transform(fireforest["month"])
```

```python
fireforest["size_category"]=le.fit_transform(fireforest["size_category"])
```

```python
fireforest
```

```python
x=fireforest.iloc[:,:11]
```

```python
y=fireforest.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential,load_model
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model=Sequential()
```

```python
model.add(layers.Dense(50,input_dim=11,activation='relu'))
```

```python
model.add(layers.Dense(1,activation="relu"))
```

```python
model.add(layers.Dense(1,activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy",optimizer='adam',metrics=['acc'])
```

```python
history=model.fit(x,y,validation_split=0.33,epochs=100,batch_size=10)
```

```python
scores=model.evaluate(x,y)
```

```python
print("%s: %2f%%" % (model.metrics_names[1],scores[1]*100))
```
